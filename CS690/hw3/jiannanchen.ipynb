```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
from pandas import DataFrame as df

from bokeh.core.properties import field
from ipywidgets import interact
from bokeh.io import push_notebook, output_notebook
from bokeh.layouts import layout
from bokeh.models import (
    ColumnDataSource, HoverTool, SingleIntervalTicker, Slider, Button, Label,
    CategoricalColorMapper, CustomJS,
)
from bokeh.palettes import Spectral6, brewer
from bokeh.plotting import figure, show
from bokeh.layouts import column, widgetbox, row, gridplot
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from bkcharts import Bar, output_file, show

output_notebook()
```

```python
def process_data_all():
    raw_data = df.from_csv("Wholesale customers data.csv")
    
    raw_data = raw_data.reset_index()

    return raw_data
```

```python
ws_df = process_data_all()

cate = list(ws_df.columns)
index = list(ws_df.index)

origin_channels = ws_df.loc[:, cate[0]].tolist()
origin_regions = ws_df.loc[:, cate[1]].tolist()

origin_channels_dict = dict()
origin_regions_dict = dict()

for c in origin_channels:
    if c not in origin_channels_dict:
        origin_channels_dict[c] = 1
    else:
        origin_channels_dict[c] += 1

for c in origin_regions:
    if c not in origin_regions_dict:
        origin_regions_dict[c] = 1
    else:
        origin_regions_dict[c] += 1

ws_df_v = ws_df.loc[:, cate[2:]]
df_values = ws_df_v.values

kmeans_channel = KMeans(n_clusters=2, random_state=0).fit(df_values)
kmeans_regions = KMeans(n_clusters=3, random_state=0).fit(df_values)

km_channel_labels = kmeans_channel.labels_
km_regions_labels = kmeans_regions.labels_

agg_channel = AgglomerativeClustering(n_clusters=2, linkage='ward').fit(df_values)
agg_regions = AgglomerativeClustering(n_clusters=3, linkage='ward').fit(df_values)

agg_channel_labels = agg_channel.labels_
agg_regions_labels = agg_regions.labels_

km_channel_dict = dict()
km_regions_dict = dict()

agg_channel_dict = dict()
agg_regions_dict = dict()

for c in km_channel_labels:
    if c not in km_channel_dict:
        km_channel_dict[c] = 1
    else:
        km_channel_dict[c] += 1
        
for c in km_regions_labels:
    if c not in km_regions_dict:
        km_regions_dict[c] = 1
    else:
        km_regions_dict[c] += 1
        
for c in agg_channel_labels:
    if c not in agg_channel_dict:
        agg_channel_dict[c] = 1
    else:
        agg_channel_dict[c] += 1
        
for c in agg_regions_labels:
    if c not in agg_regions_dict:
        agg_regions_dict[c] = 1
    else:
        agg_regions_dict[c] += 1
        
channel_cols = ['Channel', 'Count']
regions_cols = ['Regions', 'Count']

km_channel_df = pd.DataFrame.from_records(list(km_channel_dict.items()), columns=channel_cols)
km_regions_df = pd.DataFrame.from_records(list(km_regions_dict.items()), columns=regions_cols)

agg_channel_df = pd.DataFrame.from_records(list(agg_channel_dict.items()), columns=channel_cols)
agg_regions_df = pd.DataFrame.from_records(list(agg_regions_dict.items()), columns=regions_cols)

p_km_cl = Bar(km_channel_df, 'Channel', values='Count', title="Channels Clustered by KMeans", plot_width=250, plot_height=250)
p_km_rg = Bar(km_regions_df, 'Regions', values='Count', title="Regions Clustered by KMeans", plot_width=250, plot_height=250)

p_agg_cl = Bar(km_channel_df, 'Channel', values='Count', title="Channels Clustered by Hierarchical Clustering", plot_width=250, plot_height=250)
p_agg_rg = Bar(km_regions_df, 'Regions', values='Count', title="Regions Clustered by Hierarchical Clustering", plot_width=250, plot_height=250)

origin_channel_df = pd.DataFrame.from_records(list(origin_channels_dict.items()), columns=channel_cols)
origin_regions_df = pd.DataFrame.from_records(list(origin_regions_dict.items()), columns=regions_cols)

p_cl = Bar(origin_channel_df, 'Channel', values='Count', title="Original Channels", plot_width=250, plot_height=250)
p_rg = Bar(origin_regions_df, 'Regions', values='Count', title="Original Regions", plot_width=250, plot_height=250)

grid = gridplot([[p_cl, p_km_cl, p_agg_cl], [p_rg, p_km_rg, p_agg_rg]])
```

### Clustering with KMeans Clustering and Agglomerative Clustering

We can see from below that how different the clustered results are from the original clusters. From the results, we can see that the difference between these two clustering results' performance is not obvious. However, the difference between the clusters' results and the original data is huge.

```python
def int2str(l):
    res = []
    for i in l:
        res.append(str(i))
    return res
```

```python
ws_df = process_data_all()

cate = list(ws_df.columns)
index = list(ws_df.index)

xaxis_name = cate[2]
yaxis_name = cate[4]

cluster = KMeans(n_clusters=2, random_state=0, init='k-means++').fit(df_values)
cluster_labels = int2str(list(cluster.labels_))

source_cluster = ColumnDataSource(dict(x = ws_df.loc[:, xaxis_name].tolist(), y = ws_df.loc[:, yaxis_name].tolist(), legend_names = cluster_labels))

plot_cluster = figure(title=' '.join([yaxis_name, 'vs. ', xaxis_name, 'Clustered with KMeans']), plot_height=400, plot_width= 490)
plot_cluster.xaxis.ticker = SingleIntervalTicker(interval=20000)
plot_cluster.xaxis.axis_label = xaxis_name
plot_cluster.yaxis.ticker = SingleIntervalTicker(interval=20000)
plot_cluster.yaxis.axis_label = yaxis_name

color_mapper_cluster = CategoricalColorMapper(palette=brewer['Spectral'][8], factors=['0', '1', '2', '3', '4', '5', '6', '7'])

r_cluster = plot_cluster.circle(
    'x',
    'y',
    size=10,
    source=source_cluster,
    fill_color={'field': 'legend_names', 'transform': color_mapper_cluster},
    fill_alpha=0.7,
    line_color='#7c7e71',
    line_width=0.5,
    line_alpha=0.5,
    legend='legend_names',
)

plot_cluster.add_tools(HoverTool(tooltips=[(xaxis_name, '$x'), (yaxis_name, '$y')], show_arrow=True, point_policy='follow_mouse'))

def update_notebook(n_cluster=2, init='k-means++', x='Fresh', y='Grocery'):
    n_cluster = int(n_cluster)
    cluster = KMeans(n_clusters=n_cluster, random_state=0, init=init).fit(df_values)
    cluster_labels = int2str(list(cluster.labels_))
    r_cluster.data_source.data['x'] = ws_df.loc[:, x].tolist()
    r_cluster.data_source.data['y'] = ws_df.loc[:, y].tolist()
    r_cluster.data_source.data['legend_names'] = cluster_labels
    plot_cluster.title.text = ' '.join([y, 'vs.', x, 'Clustered with KMeans'])
    plot_cluster.xaxis.axis_label = x
    plot_cluster.yaxis.axis_label = y
    push_notebook()
```

### Clustering with KMeans Clustering

We can see from below that the clustering result would be affected by two factors, the number of clusters and the method for initialization. The number of clusters determines how many clusters are there in the result. The initialization method would determine that how the cluster chooses initial centroids.

* k-means++ : selects initial cluster centers for k-mean clustering in a smart way to speed up convergence. See section Notes in k_init for more details.
* random: choose k observations (rows) at random from data for the initial centroids.

```python
show(plot_cluster, notebook_handle = True)
```

```python
interact(update_notebook, n_cluster = ['2', '3', '5', '8'], init = ['k-means++', 'random'], x = cate[2:], y = cate[2:])
```

```python
ws_df = process_data_all()

cate = list(ws_df.columns)
index = list(ws_df.index)

xaxis_name = cate[2]
yaxis_name = cate[4]

cluster = AgglomerativeClustering(n_clusters=2, linkage='ward').fit(df_values)
cluster_labels = int2str(list(cluster.labels_))

source_cluster = ColumnDataSource(dict(x = ws_df.loc[:, xaxis_name].tolist(), y = ws_df.loc[:, yaxis_name].tolist(), legend_names = cluster_labels))

plot_cluster = figure(title=' '.join([yaxis_name, 'vs. ', xaxis_name, 'Clustered with Agglomerative Clustering']), plot_height=400, plot_width= 490)
plot_cluster.xaxis.ticker = SingleIntervalTicker(interval=20000)
plot_cluster.xaxis.axis_label = xaxis_name
plot_cluster.yaxis.ticker = SingleIntervalTicker(interval=20000)
plot_cluster.yaxis.axis_label = yaxis_name

color_mapper_cluster = CategoricalColorMapper(palette=brewer['Spectral'][8], factors=['0', '1', '2', '3', '4', '5', '6', '7'])

r_cluster = plot_cluster.circle(
    'x',
    'y',
    size=10,
    source=source_cluster,
    fill_color={'field': 'legend_names', 'transform': color_mapper_cluster},
    fill_alpha=0.7,
    line_color='#7c7e71',
    line_width=0.5,
    line_alpha=0.5,
    legend='legend_names',
)

plot_cluster.add_tools(HoverTool(tooltips=[(xaxis_name, '$x'), (yaxis_name, '$y')], show_arrow=True, point_policy='follow_mouse'))

def update_notebook(n_cluster=2, linkage='ward', x='Fresh', y='Grocery'):
    n_cluster = int(n_cluster)
    cluster = AgglomerativeClustering(n_clusters=n_cluster, linkage=linkage).fit(df_values)
    cluster_labels = int2str(list(cluster.labels_))
    r_cluster.data_source.data['x'] = ws_df.loc[:, x].tolist()
    r_cluster.data_source.data['y'] = ws_df.loc[:, y].tolist()
    r_cluster.data_source.data['legend_names'] = cluster_labels
    plot_cluster.title.text = ' '.join([y, 'vs.', x, 'Clustered with Agglomerative Clustering'])
    plot_cluster.xaxis.axis_label = x
    plot_cluster.yaxis.axis_label = y
    push_notebook()
```

### Clustering with Agglomerative Clustering

Agglomerative Clustering is a kind of hierachical cluster. We can see from below that the clustering result would be affected by two factors, the number of clusters and the linkage criterion. The number of clusters determines how many clusters are there in the result. The linkage would determine which distance to use between sets of observation. 
* Ward minimizes the sum of squared differences within all clusters. It is a variance-minimizing approach and in this sense is similar to the k-means objective function but tackled with an agglomerative hierarchical approach.
* Maximum or complete linkage minimizes the maximum distance between observations of pairs of clusters.
* Average linkage minimizes the average of the distances between all observations of pairs of clusters.
